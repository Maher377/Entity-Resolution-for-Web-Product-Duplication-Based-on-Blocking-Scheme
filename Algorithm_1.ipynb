# Algorithm 1: Data cleaning and shop dictionary summairzing

In algorithm 1, we propose a basic data cleaning, and summarize all the keys and values of each weh shop into a shop_key_dictionary.

```python
# First, convert the json into python dictionary
import json
    
## read the json file
f = open('C:\\\\Users\\\\mingh\\\\Desktop\\\\text\\\\Master\\\\Thesis\\\\Previous code\\\\APFAProject-master\\\\APFA_software\\\\data\\\\TVs-all-merged.json')
data = json.load(f)
    
## extract the product ID and convert it into a list
product_ID = list(data.keys())

# create dictionary of each shop to process the key value pairs
shop_dict = {}
for product in data.keys():
    data[product] = data[product][0]
for id in data.keys():
    if data[id]['shop'] in list(shop_dict.keys()):
        pass
    else:
        shop_dict[data[id]['shop']] = {}
        
# loops to record all the key names in each shop",

for id in data.keys():
    for shop_name in shop_dict.keys():
        if data[id]['shop'] == shop_name:
            if id in shop_dict[shop_name].keys() :
                pass
            else: 
                shop_dict[shop_name]['product_detail'] = dict(data[id])
        else: pass    
        
```

Then, we adjust the format of original data set, tidy up the product description into KVPs format.

```python
for shop in shop_dict.values():
    shop['product_detail']['featuresMap'] = {k: [v] for k, v in shop['product_detail']['featuresMap'].items()}
for id in data.keys():
    data[id]['featuresMap'] = {k: [v] for k, v in data[id]['featuresMap'].items()}
```

After shop dictionary is created, now we import the KVPs data into the shop dictionary.

```python
#combine the two data set
for shop in shop_dict.values():
    for id in data.keys():
        dic1 = shop['product_detail']['featuresMap']
        dic2 = data[id]['featuresMap']
        dic3 = dict(dic1)

        for k, v in dic2.items():
            dic3[k] = dic3[k] + v if k in dic3 else v
            
        shop['product_detail']['featuresMap'] = dict(dic3)
```

In case there are duplicated key values, we propose a method to clean the duplicated values.

```python
#check duplication 
for shop in shop_dict.values():
    for key in shop.keys():
        for features in shop[key]["featuresMap"].keys():
            output = []
            for x in shop[key]["featuresMap"][features]:
                if x not in output:
                    output.append(x)
            shop[key]["featuresMap"][features] = list(output)
```

In the end, we export the shop dictionary for the further algorithm.

```python
import json
with open('shop_dict.json', 'w') as f:
    json.dump(shop_dict, f)
```
